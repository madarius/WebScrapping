# Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

```python
main_url='https://contentservice.kz/catalog/'
```

```python
#get main page
req=requests.get(main_url)
soup = BeautifulSoup(req.content,'lxml')
print(soup)
```

```python
#obtain the list of products page
products_list=soup.find_all("ul", {"class": "navbar-links"})[0]
print(products_list)
```

```python
#obtain categories' links
categories=[]
for a in products_list('a', href=True):
    string=('https://contentservice.kz'+a['href'])
    categories.append(string)
    print ('https://contentservice.kz'+a['href'])
```

```python
#obtain the link of subcategories
subcategories=[]
for i in range(0,len(categories)):
    req=requests.get(categories[i])
    soup = BeautifulSoup(req.content,'lxml')
    link=soup.find_all("ul", {"class": "link-list link-list_md"})[0]
    for a in link('a', href=True):
        string='https://contentservice.kz'+a['href']
        subcategories.append(string)
        print(string)
```

```python
# A function to get the number of pages for link.
def get_page_number_test(url):
    try:
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        ul = soup.find('ul', class_='pagination')
        return int(ul.find_all('a')[-2].getText())
    except AttributeError:
        return 1
```

```python
#testing the function
get_page_number_test('https://contentservice.kz/yogurt-1/')
```

```python
# get link for each page 
def go_each_page_test(url, page_number):
    #print(url) #the first page has not word 'page inside the link'
    for i in range (2, page_number+1):
        link=url+'page'+str(i)+'/'
        subcategories.append(link)
```

```python
for sub in subcategories:
    go_each_page_test(sub,get_page_number_test(sub))
```

```python
products_links=[]
```

```python
#obtain
def obtain_products(url):
    req=requests.get(url)
    soup = BeautifulSoup(req.content,'html.parser')
    page=soup.find_all("div", {"class": "products-slider__item__name max-3-lines"})
    for i in range(0,len(page)):
        for a in page[i].find_all("a",href=True):
            #print('https://contentservice.kz'+a['href'])
            products_links.append('https://contentservice.kz'+a['href'])
```

```python
#all the products links was obtained and stored in the list
for i in range (0,len(subcategories)):
    obtain_products(subcategories[i])
```

```python
import csv
```

```python
filename = 'products.csv'
with open(filename, 'w') as f:
    w = csv.writer(f)
    w.writerow(['Category', 'Subcategory', 'Title', 'TM', 'Country'])
```

```python
def parse_page(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
    ol=soup.find('ol',class_='breadcrumb')
    category=ol.find_all('a')
    category1=category[-2].getText()
    category2=category[-1].getText()
    div = soup.find('section', class_='site-content')
    title = div.find('h1').getText().strip()
    tm ='TM '+div.find('ul', class_='list-unstyled list-lg-pad').find('a').getText()
    country=div.find_all('span',title=True)[0]['title']
    with open(filename, 'a',encoding='utf-8') as f:
        w = csv.writer(f)
        w.writerow([category1,category2, title, tm, country])
```

```python
for product in products_links:
    try:
        parse_page(product)
    except IndexError:
        response = requests.get(product)
        soup = BeautifulSoup(response.text, 'html.parser')
        ol=soup.find('ol',class_='breadcrumb')
        category=ol.find_all('a')
        category1=category[-2].getText()
        category2=category[-1].getText()
        div = soup.find('section', class_='site-content')
        title = div.find('h1').getText().strip()
        tm='TM '+div.find('ul', class_='list-unstyled list-lg-pad').find('a').getText()
        with open(filename, 'a',encoding='utf-8') as f:
            w = csv.writer(f)
            w.writerow([category1,category2, title, tm, None])
    except AttributeError:
        try:
            response = requests.get(product)
            soup = BeautifulSoup(response.text, 'html.parser')
            ol=soup.find('ol',class_='breadcrumb')
            category=ol.find_all('a')
            category1=category[-2].getText()
            category2=category[-1].getText()
            div = soup.find('section', class_='site-content')
            title = div.find('h1').getText().strip()
            country=div.find_all('span',title=True)[0]['title']
            with open(filename, 'a',encoding='utf-8') as f:
                w = csv.writer(f)
                w.writerow([category1,category2, title, None, country])
        except IndexError:
            response = requests.get(product)
            soup = BeautifulSoup(response.text, 'html.parser')
            ol=soup.find('ol',class_='breadcrumb')
            category=ol.find_all('a')
            category1=category[-2].getText()
            category2=category[-1].getText()
            div = soup.find('section', class_='site-content')
            title = div.find('h1').getText().strip()
            with open(filename, 'a',encoding='utf-8') as f:
                w = csv.writer(f)
                w.writerow([category1,category2, title, None, country])
```
